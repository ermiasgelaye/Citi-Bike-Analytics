```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
file_name1 = os.path.join("raw_data/201801-citibike-tripdata.csv")
file_name2 = os.path.join("raw_data/201802-citibike-tripdata.csv")
file_name3 = os.path.join("raw_data/201803-citibike-tripdata.csv")
file_name4 = os.path.join("raw_data/201804-citibike-tripdata.csv")
file_name5 = os.path.join("raw_data/201805-citibike-tripdata.csv")
file_name6 = os.path.join("raw_data/201806-citibike-tripdata.csv")
file_name7 = os.path.join("raw_data/201807-citibike-tripdata.csv")
file_name8 = os.path.join("raw_data/201808-citibike-tripdata.csv")
file_name9 = os.path.join("raw_data/201809-citibike-tripdata.csv")
file_name10 = os.path.join("raw_data/201810-citibike-tripdata.csv")
file_name11 = os.path.join("201801-citibike-tripdata.csv")
file_name12 = os.path.join("raw_data/JC-201712-citibike-tripdata.csv")

january_trip_data = pd.read_csv(file_name1)
february_trip_data = pd.read_csv(file_name2)
march_trip_data = pd.read_csv(file_name3)
april_trip_data = pd.read_csv(file_name4)
may_trip_data = pd.read_csv(file_name5)
june_trip_data = pd.read_csv(file_name6)
july_trip_data = pd.read_csv(file_name7)
august_trip_data = pd.read_csv(file_name8)
september_trip_data = pd.read_csv(file_name9)
october_trip_data = pd.read_csv(file_name10)
november_trip_data = pd.read_csv(file_name11)
december_trip_data = pd.read_csv(file_name12)
```

```python
#merge tabels
framesII = [january_trip_data, february_trip_data, march_trip_data, april_trip_data,
           may_trip_data, june_trip_data, july_trip_data, august_trip_data,
           september_trip_data, october_trip_data, november_trip_data, december_trip_data]
```

```python
resultII = pd.concat(framesII)
```
