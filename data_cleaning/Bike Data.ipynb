```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
from datetime import datetime
```

```python
file_name1 = os.path.join("raw_data/JC-201909-citibike-tripdata.csv")
file_name2 = os.path.join("raw_data/JC-201910-citibike-tripdata.csv")
file_name3 = os.path.join("raw_data/JC-201911-citibike-tripdata.csv")
file_name4 = os.path.join("raw_data/JC-201912-citibike-tripdata.csv")
file_name5 = os.path.join("raw_data/JC-202001-citibike-tripdata.csv")
file_name6 = os.path.join("raw_data/JC-202002-citibike-tripdata.csv")
file_name7 = os.path.join("raw_data/JC-202003-citibike-tripdata.csv")
file_name8 = os.path.join("raw_data/JC-202004-citibike-tripdata.csv")
file_name9 = os.path.join("raw_data/JC-202005-citibike-tripdata.csv")
file_name10 = os.path.join("raw_data/JC-202006-citibike-tripdata.csv")
file_name11 = os.path.join("raw_data/JC-202007-citibike-tripdata.csv")
file_name12 = os.path.join("raw_data/JC-202008-citibike-tripdata.csv")

september19_trip_data = pd.read_csv(file_name1)
october19_trip_data = pd.read_csv(file_name2)
november19_trip_data = pd.read_csv(file_name3)
december19_trip_data = pd.read_csv(file_name4)
january20_trip_data = pd.read_csv(file_name5)
february20_trip_data = pd.read_csv(file_name6)
march20_trip_data = pd.read_csv(file_name7)
april20_trip_data = pd.read_csv(file_name8)
may20_trip_data = pd.read_csv(file_name9)
june20_trip_data = pd.read_csv(file_name10)
july20_trip_data = pd.read_csv(file_name11)
august20_trip_data = pd.read_csv(file_name12)
```

```python
january20_trip_data.head()
```

```python
#merge tabels
frames= [september19_trip_data, october19_trip_data, november19_trip_data, december19_trip_data,
           january20_trip_data, february20_trip_data, march20_trip_data, april20_trip_data,
           may20_trip_data, june20_trip_data, july20_trip_data, august20_trip_data]
```

```python
result = pd.concat(frames)
```

```python
rename_df = result.rename(columns={
    "tripduration": "trip_duration",
    "starttime": "start_time",
    "stoptime": "stop_time",
    "start station id": "start_station_id",
    "start station name": "start_station_name",
    "start station latitude": "start_station_latitude",
    "start station longitude": "start_station_longitude",
    "end station id": "end_station_id",
    "end station name": "end_station_name",
    "end station latitude": "end_station_latitude",
    "end station longitude": "end_station_longitude",
    "bikeid": "bike_id",
    "usertype": "user_type",
    "birth year": "birth_year"
})
```

```python
# Change the data type to string forid's
rename_df["start_station_id"] = rename_df["start_station_id"].astype(str)
rename_df["end_station_id"] = rename_df["end_station_id"].astype(str)
rename_df["bike_id"] = rename_df["bike_id"].astype(str)
```

```python
# lable the one and zero values to gender type
rename_df["gender"] = rename_df["gender"].replace({
    0: "Unknown",
    1: "Male",
    2: "Female"
})
rename_df
```

```python
# get age of the riders
currentYear = datetime.now().year

rename_df["age"] = currentYear - rename_df["birth_year"]

rename_df
```

```python
rename_df.to_csv("output/2019-2020_trip_data.csv")
```
